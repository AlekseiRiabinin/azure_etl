```python
from datetime import datetime, timedelta
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from airflow.models import DAG
from airflow.operators.python import PythonOperator

# Simplified visualization setup
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (12, 4)

def mock_spark_etl(**kwargs) -> dict[str, str | float]:
    """Spark ETL simulation."""
    print("Starting Spark ETL simulation")
    
    # Generate test data
    data = pd.DataFrame({
        'value': np.random.normal(100, 15, 1000),
        'timestamp': pd.date_range(end=datetime.now(), periods=1000, freq='min')
    })
    
    # Basic visualization
    fig, (ax1, ax2) = plt.subplots(1, 2)
    
    ax1.hist(data['value'], bins=30)
    ax1.set_title("Value Distribution")
    
    hourly_counts = data.resample('H', on='timestamp').count()
    ax2.plot(hourly_counts.index, hourly_counts['value'], marker='o')
    ax2.set_title("Records Processed by Hour")
    
    plt.tight_layout()
    plt.show()
    
    return {
        "status": "success",
        "records_processed": len(data),
        "output_path": "s3://output/processed_data.parquet"
    }

# Clean DAG definition
with DAG(
    dag_id="spark_etl_simple",
    schedule="@hourly",
    start_date=datetime(2024, 1, 1),
    catchup=False,
    default_args={
        "retries": 2,
        "retry_delay": timedelta(minutes=3)
    }
) as dag:
    
    process_task = PythonOperator(
        task_id="process_data",
        python_callable=mock_spark_etl
    )

# Development testing
print("Development Test Run:")
test_result = mock_spark_etl()
print("\nTest Results:")
for k, v in test_result.items():
    print(f"{k:>18}: {v}")

# DAG info
print("\nDAG Configuration:")
print(f"{'DAG ID':>18}: {dag.dag_id}")
print(f"{'Schedule':>18}: {dag.schedule_interval}")
print(f"{'Start Date':>18}: {dag.start_date.strftime('%Y-%m-%d')}")
print(f"{'Tasks':>18}: {[t.task_id for t in dag.tasks]}")
```
