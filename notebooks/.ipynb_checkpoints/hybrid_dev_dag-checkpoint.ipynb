```python
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
from airflow.models import DAG
from airflow.operators.python import PythonOperator
import psycopg2
from psycopg2 import sql

# Tables from your working code
TABLES_TO_CHECK = [
    "bridge_customer_address",
    "dim_weather_station",
    "dim_customer",
    "fact_billing",
    "fact_weather_daily",
    "dim_meter",
    "fact_meter_readings",
    "dim_tariff"
]

def postgres_etl(**kwargs) -> dict[str, str | float]:
    """Postgres ETL with table validation."""
    print("Starting Postgres ETL")
    
    try:
        # Database connection
        conn = psycopg2.connect(
            host="postgres",
            database="postgres",
            user="postgres",
            password="postgres",
            port="5432"
        )
        print(f"Connected to {conn.info.dbname}")
        
        results = {}
        with conn.cursor() as cur:
            # Check each table
            for table in TABLES_TO_CHECK:
                try:
                    cur.execute(
                        sql.SQL("SELECT COUNT(*) FROM {}").format(sql.Identifier(table))
                    )
                    count = cur.fetchone()[0]
                    results[table] = count
                    print(f"{table:25} {count:>8,} records")
                
                except psycopg2.Error as e:
                    print(f"{table:25} NOT FOUND")
                    continue
            
            # Visualization of available tables
            if results:
                fig, ax = plt.subplots()
                pd.DataFrame.from_dict(results, orient='index', columns=['Count'])\
                    .plot(kind='bar', ax=ax, legend=False)
                ax.set_title("Table Record Counts")
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()
                
                return {
                    "status": "success",
                    "counts": results,
                    "timestamp": datetime.now().isoformat()
                }
            else:
                raise ValueError("No valid tables found")
                
    except Exception as e:
        print(f"Pipeline failed: {e}")
        raise
    finally:
        if 'conn' in locals():
            conn.close()
            print("Connection closed")

# DAG definition
with DAG(
    dag_id="postgres_table_monitor",
    schedule="@daily",
    start_date=datetime(2024, 1, 1),
    catchup=False,
    default_args={
        "retries": 3,
        "retry_delay": timedelta(minutes=5)
    }
) as dag:
    
    monitor_task = PythonOperator(
        task_id="check_table_counts",
        python_callable=postgres_etl
    )

# Test execution
print("\nTest Execution:")
try:
    test_result = postgres_etl()
    print("\nResults Summary:")
    for table, count in test_result["counts"].items():
        print(f"{table:25} {count:>8,}")
except Exception as e:
    print(f"Test failed: {e}")

print("\nDAG Ready:", dag.dag_id)
```
